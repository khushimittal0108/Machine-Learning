```
import numpy as np
import pandas as pd
```

```
import matplotlib.pyplot as plt
import seaborn as sns
```

```
from sklearn.datasets import load_boston
```

```
boston= load_boston()
```

```
boston.keys()
```

```
print(boston['feature_names'])
```

```
boston['data']
```

```
df=pd.DataFrame(data=boston['data'],columns=boston['feature_names'])
```

```
df.head()
```

```
df.columns
```

```
print(boston['DESCR'])
```

```
df['MEDV']=boston['target']
```

```
df.head()
```

```
sns.pairplot(df)
```

```
df.columns
```

```
X=df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']]
y=df['MEDV']
```

```
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33)
```

```
from sklearn.linear_model import LinearRegression
```

```
lm=LinearRegression()
```

```
lm.fit(X_train,y_train)
```

```
prediction=lm.predict(X_test)
```

```
sns.scatterplot(x=y_test,y=prediction)
```

```
x=range(0,167)
plt.plot(x,prediction,label='Predicted')
plt.plot(x,y_test,label='Actual')
plt.legend()
```

```
sns.scatterplot(x,prediction,label='Predicted')
sns.scatterplot(x,y_test,label='Actual')
plt.legend()
```

```
sns.distplot((y_test-prediction),bins=50)
```

```
lm.coef_
```

```
coefficients=pd.DataFrame(lm.coef_,X.columns,columns=['Coefficients'])
```

```
coefficients
```

```
from sklearn import metrics
```

```
#mean absolute error
print('MAE:',metrics.mean_absolute_error(y_test,prediction))
```

```
#mean squared error 
print('MSE:',metrics.mean_squared_error(y_test,prediction))
```

```
#Root mean square error
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,prediction)))
```

```
#variance score
metrics.explained_variance_score(y_test,prediction)
```
